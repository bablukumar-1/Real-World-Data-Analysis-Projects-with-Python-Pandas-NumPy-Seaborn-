```python
# Create DataFrame
faq_df = pd.DataFrame(faq_data)

# Encode FAQ questions
faq_embeddings = model.encode(faq_df['question'].tolist())

# ques = faq_data['question']
# ans = faq_data['answer']
# print(len(ques),len( ans))
```

```python
def dynamic_faq_system():
    print("Chatbot: Hello! Ask a question or type 'exit' to quit.")
    
    while True:

        user_query = input("You: ")
        print("User Input :", user_query)
        if user_query.lower() in ["exit", "quit"]:
            print("Chatbot: Goodbye!")
            break
        
        
        # Encode user query
        query_embedding = model.encode(user_query)
        # print('embeded query: ', query_embedding)
        # Find the closest FAQ question
        closest = util.semantic_search(query_embedding, faq_embeddings, top_k=1)
        best_match_id = closest[0][0]['corpus_id']
        best_answer = faq_df.iloc[best_match_id]['answer']
        
        score = closest[0][0]['score']   
        if score < 0.4:
           print("Chatbot: Sorry, I don't know the answer.")
        else:
           print("Chatbot:", best_answer)
        # Display answer
        # print(f"Chatbot: {best_answer}")
```

```python
# Run the FAQ chatbot
dynamic_faq_system()
```

```python
def trans(text, target):
    trans_text = gt(target=target).translate(text)
    return trans_text
```

```python
while True:
    print("\n Language Translator Tool")
    input_text = input("Enter Your Text :")
    if input_text.lower()=='exit':
        break
    target_lang=input("To translate language (en,hi,banagali...)")
    trans_text = trans(input_text, target_lang)
    print('result', trans_text)
```
