```python
def extract_top_keyword(content, keywords,keyword_embeddings, top_n=5):
    keyword_embedding = model.encode(keywords)
    content_embedding = model.encode(content)

    similarities = util.cos_sim(content_embedding, keyword_embedding).flatten()


    top_indices = torch.topk(similarities,top_n)

    top_keywords = [keywords[i] for i in top_indices]

    return top_keywords
    
```

```python
def extract_top_keyword(content, keywords, top_n=5):
    keyword_embeddings = model.encode(keywords)
    content_embedding = model.encode(content)

    similarities = util.cos_sim(content_embedding, keyword_embeddings).flatten()

    # Correct unpacking
    _, top_indices = torch.topk(similarities, top_n)

    top_keywords = [keywords[i] for i in top_indices.tolist()]

    return top_keywords
```
