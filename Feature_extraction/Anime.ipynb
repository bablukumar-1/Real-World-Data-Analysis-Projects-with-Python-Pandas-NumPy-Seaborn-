```python
import pandas as pd
import numpy as np
```

## Loading Data

```python
df = pd.read_csv(r'anime.csv')
df.head()
```

```python
## Make a new column for episode count

df.loc[2]['Title']
```

```python
def extract_episode(txt):
    check = False
    data = ""
    for i in txt:
        if i==")":
            check=False
            return data

        if check == True:
            data = data+i
        if i=="(":
            check=True
    
```

```python
df["Episode"] = df['Title'].apply(extract_episode)
df['Episode']=df['Episode'].str.replace(" eps","")
```

```python
df['Episode']=df['Episode'].astype(int)
df.head()
```

## Make a new column for time stamp

```python
df.loc[0]['Title']
```

```python
def extraction_time(txt):
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            for j in range(i+1,i+20):
                data +=txt[j]
            return data
        
```

```python
df["Total Time"]=df['Title'].apply(extraction_time)
```
